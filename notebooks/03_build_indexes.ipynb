# Build FAISS Indices

Build FAISS indices for efficient similarity search.

```python
import sys
sys.path.insert(0, '..')

from src.retriever import Retriever
import logging

logging.basicConfig(level=logging.INFO)
logger = logging.getLogger(__name__)
```

## Initialize Retriever

```python
retriever = Retriever(index_path='../data/')
print("Retriever initialized")
```

## Build Indices for Categories

```python
categories = ['bottle', 'carpet', 'wood', 'cable', 'capsule']

for category in categories:
    print(f"Building index for {category}...")
    retriever.load_index(category)
    print(f"Index built for {category}")

print("All indices built successfully")
```
